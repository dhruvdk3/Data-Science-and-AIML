pd.Series() 
Create a Series from a list, array, or dictionary

pd.DataFrame() 
Create a DataFrame from a list of lists, arrays, or dictionaries
it has 4 arguments
data, index, columns, dtype 
out of these data is mandatory
data is a list of lists, arrays, or dictionaries default is None
index is a list of row labels default is None
columns is a list of column labels default is None
dtype is the data type of the data default is None


```python
pd.Series([1,2,3,4,5])
pd.Series([1,2,3,4,5], index=[1,2,3,4,5])
pd.Series([1,2,3,4,5], index=[1,2,3,4,5], dtype='float64')
pd.Series([1,2,3,4,5], index=[1,2,3,4,5], dtype='float64', name='numbers')
```

df.describe() provides a summary of the DataFrame like mean, std, min, max, count, etc.
we can also use describe on a specific column
like df.age.describe()
if we need describe multiple columns we can use df[['age', 'city']].describe()


to generate summery for non numerical columns we can use df.describe(include=['O'])

we can describe all the data as df.describe(include='all')


pandas calculate percentile at 25 50 75 position 
but we can coustomize it as 
df.describe(percentiles=[0.2,0.4,0.7,0.8])

```python
data = {'name': ['Alice', 'Bob', 'Charlie'], 'age': [25, 30, 35], 'city': ['New York', 'Los Angeles', 'Chicago'], 'salary': [100000, 120000, 130000]}
df = pd.DataFrame(data)
# df.describe()
df[['age', 'salary']].describe()
df.describe(include = 'object')
df.describe(percentiles=[0.2,0.4,0.7,0.8])
```

```python
salary = pd.read_csv("https://github.com/YBI-Foundation/Dataset/raw/refs/heads/main/Salary%20Data.csv")
salary
```

```python
salary.info() 
salary.shape
```

```python
Y = salary['Salary']
X = salary[['Experience Years']]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, train_size=0.7, random_state=2529)
```

step 1 : Import Libery
step 2 : Import Dataset
step 3 : Define Target
step 4 : Split
step 5 : Select Model
step 6 : Training Model (Fit model)
step 7 : Prediction
step 8 : Accuracy

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

now model is
y = intercept(26596) + coefficent(9405)*Experience + error

```python
y_predicted = model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_percentage_error
mean_absolute_percentage_error(y_test, y_predicted)
```
